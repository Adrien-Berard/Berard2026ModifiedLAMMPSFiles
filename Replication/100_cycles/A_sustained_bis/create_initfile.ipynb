```python
import numpy as np

def write_initial_file():
    with open('InitialFile.txt', 'w') as f:
        f.write('# Initialisation file of the particles\n\n')
        n_monomers = 80
        n_HP1 = 400
        n_Swi6M = 0
        n_ghost = 0

        n_chromatin_bonds = n_monomers - 1
        n_angles = n_monomers - 2
        n_dihedral = n_angles - 1

        f.write(f'{n_monomers + n_HP1 + n_Swi6M + n_ghost}  atoms\n')
        f.write(f'{n_chromatin_bonds}  bonds\n\n')
        #f.write(f'{n_angles}  angles\n')
        #f.write(f'{n_dihedral}  dihedrals\n\n')

        f.write('5  atom types\n')
        f.write('1  bond types\n\n')

        f.write('0.0000   50.0000 xlo xhi\n')
        f.write('0.0000   50.0000 ylo yhi\n')
        f.write('0.0000   50.0000 zlo zhi\n\n')

        f.write('Atom Type Labels\n\n')
        f.write('1    A\n')
        f.write('2    U\n')
        f.write('3    M\n')
        f.write('4    Swi6\n')
        f.write('5    Swi6M\n\n')

        f.write('Bond Type Labels\n\n')
        f.write('1    Normal\n\n')
        
        f.write('Masses\n\n')
        f.write('1    1.00 \n')
        f.write('2    1.00 \n')
        f.write('3    1.00 \n')
        f.write('4    1.00 \n')
        f.write('5    1.00 \n\n')

        monomer_types = np.random.choice(["A","U","M"], n_monomers, p=[0.9, 0.1, 0.0])

        f.write('Atoms\n\n')

        # Generate positions for monomers
        for i in range(n_monomers):
            x = i * 0.6 + 10
            y = i * 0.6 + 10
            z = i * 0.6 + 10
            f.write(f'  {i + 1}  {i + 1}  {monomer_types[i]}  {x}  {y}  {z}\n')

        # Generate positions for HP1
        hp1_positions = np.random.uniform(1, 50, (n_HP1, 3))
        for i, (x, y, z) in enumerate(hp1_positions):
            f.write(f'  {n_monomers + i + 1}  {n_monomers + i + 1}  Swi6  {x}  {y}  {z}\n')

        # Generate positions for Swi6M
        swi6m_positions = np.random.uniform(1, 50, (n_Swi6M, 3))
        for i, (x, y, z) in enumerate(swi6m_positions):
            f.write(f'  {n_monomers + n_HP1 + i + 1}  {n_monomers + n_HP1 + i + 1}  Swi6M  {x}  {y}  {z}\n')

        # Generate positions for ghost particles  
        x, y, z = np.linspace(1, 50, 50), np.linspace(1, 50, 50), np.linspace(1, 50, 50)
        x, y, z = np.meshgrid(x, y, z)
        x, y, z = x.flatten(), y.flatten(), z.flatten()
        
        for i in range(n_ghost):
            f.write(f'  {n_monomers + n_HP1 + n_Swi6M + i + 1}  {n_monomers + n_HP1 + n_Swi6M + i + 1}  G  {x[i]}  {y[i]}  {z[i]}\n')

        f.write('\nBonds\n\n')
        for i in range(n_chromatin_bonds):
            f.write(f'  {i + 1}  1  {i + 1}  {i + 2}\n')

write_initial_file()
```
